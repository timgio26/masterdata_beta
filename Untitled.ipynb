```python
# import requests
# import json
import pandas as pd
resdf=pd.DataFrame(index=[])
# [125529855,125529818,125529759,125529465,125529327]
for order in ordlist:
    print(order)
    url = f"https://accounts.forstok.com/api/v1/orders/{order}.json"
    payload = {'user_id': 'grace.permatasari@nutrifood.co.id', 'auth_type': 'jwt'}
    headers = {'Content-Type': 'application/json',"Authorization": f"Bearer {token}"}
    response = requests.request("GET", url, headers=headers, params=payload)
    
    df=pd.DataFrame(response.json()['item_lines'])
    if len(response.json()['item_lines'])>0:
        df['Qty. Invoiced']=df.groupby('sku').cumcount()+1
        if len(df.index)>1:
            print('lebih dr 1 pcs')
        df=df.drop_duplicates('sku',keep='last').rename(columns={'name':'product_name'})
        df['order_id']=order

        for i in response.json().items():
            # print(i[0])
            if i[0] in ['local_name','status','store_name','channel','ordered_at','subtotal','discount_amount','shipping_price','total_price']:
                df[i[0]]=f"{i[1]}"
                # display(df[[i[0]]])
            elif i[0]=='address':
                df['address_1']=f"{response.json()['address']['address_1']}"
                for data in ['city','country','name','phone','postal_code']:
                    df[f'{data}']=f"{response.json()['address'][f'{data}']}"
            # print()
        df=df[['order_id','id','local_name','status','store_name','channel',
            'ordered_at','product_name','sku','price',
            'sale_price','Qty. Invoiced','subtotal','total_price',
            'discount_amount','shipping_provider','shipping_price','total_price',
            'name','phone','address_1','city','country','postal_code']]
        resdf=resdf.append(df).reset_index(drop=True)
    else:
        print('cancel no item')
```
